```python
from langchain_community.tools.sql_database.tool import (
    QuerySQLDataBaseTool,
    InfoSQLDatabaseTool,
    ListSQLDatabaseTool,
    QuerySQLCheckerTool
)
from langchain.tools import tool

@tool("list_tables")
def list_tables() -> str:
    """Lista todas as tabelas disponíveis no banco de dados auditoria.db."""
    return ListSQLDatabaseTool(db=db).invoke("")


@tool("tables_schema")
def tables_schema(tables: str) -> str:
    """Retorna o schema e amostras de registros das tabelas especificadas (ex: 'apontamentos, planos_acao')."""
    return InfoSQLDatabaseTool(db=db).invoke(tables)


@tool("execute_sql")
def execute_sql(sql_query: str) -> str:
    """Executa uma consulta SQL diretamente nas tabelas do banco auditoria.db."""
    return QuerySQLDataBaseTool(db=db).invoke(sql_query)


@tool("check_sql")
def check_sql(sql_query: str) -> str:
    """Verifica se a consulta SQL está correta e sugere correções se necessário."""
    return QuerySQLCheckerTool(db=db, llm=llm).invoke({"query": sql_query})
```

```python
from crewai import Agent
from textwrap import dedent
from langchain.chat_models import ChatOpenAI

llm = ChatOpenAI(model="gpt-3.5-turbo", temperature=0)  # ou gpt-3.5-turbo, se preferir

sql_dev = Agent(
    role="Engenheiro de Dados SQL",
    goal="Responder perguntas analisando os dados do banco auditoria.db",
    backstory=dedent("""
        Você é um especialista em SQL encarregado de atender pedidos de auditoria.
        Use as ferramentas disponíveis para listar as tabelas, inspecionar os esquemas - note que por vezes o nome não é trivial, então inicie buscando pelas tabelas e posteriormente os campos relevantes
        revisar queries e buscar informações relacionadas a riscos, planos de ação, auditorias e apontamentos.
        Nunca use hífens para chamar ferramentas. Sempre use os nomes exatos como: list_tables, tables_schema etc.
    """),
    llm=llm,
    tools=[list_tables, tables_schema, check_sql, execute_sql],
    allow_delegation=False,
    verbose=True
)
```

```python
from crewai import Task

task_sql_query = Task(
    description=dedent("""
        Liste os apontamentos que possuem risco relacionado à 'Segurança da informação'
        ou 'Privacidade de dados'. Use SQL para consultar a tabela 'apontamentos'.
    """),
    agent=sql_dev,
    expected_output="Lista dos apontamentos com esses riscos.",
)
```

```python
from crewai import Crew

crew = Crew(
    agents=[sql_dev],
    tasks=[task_sql_query],
    verbose=2
)

crew.kickoff()
```

# Crew

```python
llm = ChatOpenAI(model="gpt-3.5-turbo", temperature=0)  # ou gpt-3.5-turbo, se preferir

# === Conexão com banco SQLite ===
db = SQLDatabase.from_uri("sqlite:///auditoria_v2.db")

# === Tools SQL ===
@tool("list_tables")
def list_tables() -> str:
    """Lista todas as tabelas do banco auditoria.db."""
    return ListSQLDatabaseTool(db=db).invoke("")

@tool("tables_schema")
def tables_schema(tables: str) -> str:
    """Retorna o schema das tabelas."""
    return InfoSQLDatabaseTool(db=db).invoke(tables)

@tool("execute_sql")
def execute_sql(sql_query: str) -> str:
    """Executa uma query SQL contra o banco auditoria.db."""
    return QuerySQLDataBaseTool(db=db).invoke(sql_query)

@tool("check_sql")
def check_sql(sql_query: str) -> str:
    """Valida uma query SQL antes de executar."""
    return QuerySQLCheckerTool(db=db, llm=llm).invoke({"query": sql_query})

# === Agente SQL ===
sql_dev = Agent(
    role="Engenheiro de Dados SQL",
    goal="Executar queries solicitadas pelos outros agentes",
    backstory=dedent("""
        Você é um especialista em SQL encarregado de atender pedidos de auditoria.
        Use as ferramentas disponíveis para listar as tabelas, inspecionar os esquemas - note que por vezes o nome não é trivial, então inicie buscando pelas tabelas e posteriormente os campos relevantes
        Certifique-se de que as queries estão corretas antes de executá-las - considerando que você deve consultar sempre as tabelas, e os esquemas para ter certeza quanto a melhor query
        revisar queries e buscar informações relacionadas a riscos, planos de ação, auditorias e apontamentos.
        Nunca use hífens para chamar ferramentas. Sempre use os nomes exatos como: list_tables, tables_schema etc."""),
    tools=[list_tables, tables_schema, check_sql, execute_sql],
    llm=llm,
    allow_delegation=False,
    verbose=True
)

# === Agentes de escrita ===
summary_agent = Agent(
    role="Analista de Riscos",
    goal="Interpretar os testes com falha e montar a pergunta para o dev",
    backstory="Você monta a questão SQL com base no código da OS informado.",
    llm=llm,
    verbose=True
)

findings_writer_agent = Agent(
    role="Responsável pelos Apontamentos",
    goal="Interpretar os apontamentos de uma OS e escrever o parecer",
    backstory="Você escreve uma análise técnica com base nos dados extraídos pelo dev.",
    llm=llm,
    verbose=True
)

action_conclusion_agent = Agent(
    role="Especialista em Conclusão",
    goal="Interpretar os planos de ação e concluir o relatório",
    backstory="Você elabora a conclusão do relatório com base nos planos de ação executados.",
    llm=llm,
    verbose=True
)

# === Input ===
codigo_os = "OS007"

# === Tasks ===
task_gerar_query_riscos = Task(
    description=f"""
        Gere a query SQL para buscar os testes da OS {codigo_os} cujo resultado seja diferente de 'Conforme'.
        A query deve buscar da tabela 'testes'.
    """,
    agent=summary_agent,
    expected_output="Query SQL gerada para testes falhos da OS."
)

task_exec_query_riscos = Task(
    description="Use a query fornecida para extrair os testes com falha usando execute_sql.",
    agent=sql_dev,
    context=[task_gerar_query_riscos],
    expected_output="Resultado da query com testes falhos."
)

task_escrever_sumario = Task(
    description="Use os dados dos testes com falha para escrever o sumário executivo.",
    agent=summary_agent,
    context=[task_exec_query_riscos],
    expected_output="Sumário executivo dos riscos."
)

# === Crew ===
crew = Crew(
    agents=[sql_dev, summary_agent, findings_writer_agent, action_conclusion_agent],
    tasks=[
        task_gerar_query_riscos,
        task_exec_query_riscos,
        task_escrever_sumario
    ],
    verbose=2
)

crew.kickoff()
```
